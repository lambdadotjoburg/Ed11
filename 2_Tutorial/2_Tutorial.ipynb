 ### Overlapping Graphs

```python
%reset
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 50)

y = x ** 2
z = np.sqrt(x)

fig = plt.figure(figsize=(8,8))

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Main Axis - # Left, bottom, width, height of the new axes
axes2 = fig.add_axes([0.25, 0.3, 0.4, 0.3]) # Overlapping Axis - # Left, bottom, width, height of the new axes

# Main Axis
axes1.plot(x, y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('$y = x^2$')

# Overlapping Axis
axes2.plot(x, z)
axes2.set_xlabel('x')
axes2.set_ylabel('z')
axes2.set_title('$z = \sqrt{x}$')

plt.show()
```

```python
fig = plt.figure(figsize=(8,8))

fig.text(0.5, 0.95, "Title \n", fontsize=20)

# Main Axis
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Left, bottom, width, height of the new axes
# Overlapping Axis
axes2 = fig.add_axes([0.4, 0.4, 0.4, 0.4]) # Left, bottom, width, height of the new axes

# Main Axis
axes1.plot(x, y, color='green')
axes1.set_xlabel('x', loc='right')
axes1.set_ylabel('y', loc='top', rotation=0)
axes1.set_title('$y = x^2$')
axes1.set_xticks([0,1,2,3], ['A', 'B', 'C', 'D'])
axes1.set_yticks([5,10,15,20], ['$\\alpha$', '$\\ beta$', '$\\gamma$', '$\\delta$'])

# Overlapping Axis
axes2.plot(x, z, color='red')
axes2.set_xlabel('x', loc='right')
axes2.set_ylabel('z', loc='top', rotation=0)
axes2.set_title('$z = \sqrt{x}$')
axes2.set_xticks([0,1,2,3], ['a', 'b', 'c', 'd'])

plt.show()
```

$$ y = mx + c$$

```python
m = 1
c = 2

Dx = np.linspace(-3,3,100) # Specify independent variable as a domain with interval step size

# Define relationship between independent and dependent variable

def f(x):
    return m*x + c
```
