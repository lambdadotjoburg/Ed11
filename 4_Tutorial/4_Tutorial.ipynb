##### Breath Held Under Water versus Lung Capacity

```python
import random

# set seed for reproducibility
random.seed(42)

# generate lung capacity values (in liters)
lung_capacity = [random.uniform(3, 6) for _ in range(50)]

# generate time for breath held values (in seconds)
time_breath_held = [random.uniform(10, 90) for _ in range(50)]

# print first 5 values of each list for verification
print("Lung capacity:", lung_capacity[:5])
print("Time for breath held:", time_breath_held[:5])
```

```python
print(lung_capacity)
```

```python
print(time_breath_held)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8,8))

plt.plot(lung_capacity, time_breath_held, marker='o', linestyle='none')

plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.scatter(lung_capacity, time_breath_held, marker='o')

plt.show()
```

```python
sample_size = 50

plt.figure(figsize=(8,8))

dot_sizes=np.pi*(15*np.random.rand(sample_size)**2)

plt.scatter(lung_capacity, time_breath_held, marker='o', s=dot_sizes)

plt.show()
```

```python
plt.figure(figsize=(8,8))

dot_sizes=np.pi*(15*np.random.rand(sample_size)**2)
colors=np.random.rand(sample_size)

plt.scatter(lung_capacity, time_breath_held, marker='o', s=dot_sizes, c=colors)

plt.show()
```

```python
plt.figure(figsize=(8,8))

dot_sizes=np.pi*(15*np.random.rand(sample_size)**2)
colors=np.random.rand(sample_size)

plt.scatter(lung_capacity, time_breath_held, marker='o', s=dot_sizes, c=colors)

plt.xlabel(r'Lung Capacity $[litres]$', fontsize=15)
plt.ylabel(r'Time Breath Held $[s]$', fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(8,8))

dot_sizes=np.pi*(15*np.random.rand(sample_size)**2)
colors=np.random.rand(sample_size)

plt.scatter(lung_capacity, time_breath_held, marker='o', s=dot_sizes, c=colors)

plt.xlabel(r'Lung Capacity $[litres]$', fontsize=10)
plt.ylabel(r'Time Breath Held $[s]$', fontsize=10)

plt.title('Graph Representing Time Breath Held Under Water versus Lung Capacity \n', fontsize=10)

plt.savefig('graph_representing_time_breath_helde_under_water_versus_lung_capacity.jpg')

plt.show()
```

#### Photosynthesis versus Exposure to Sunlight

```python
   import random

# set seed for reproducibility
random.seed(42) 

# generate exposure to sunlight values (in hours)
sunlight_exposure = [random.uniform(2, 8) for _ in range(50)]

# generate photosynthesis values (in arbitrary units)
photosynthesis = [sunlight * random.uniform(0.5, 1.5) for sunlight in sunlight_exposure]

# print first 5 values of each list for verification
print("Sunlight exposure:", sunlight_exposure[:5])
print("Photosynthesis:", photosynthesis[:5]) 
```

```python
print(sunlight_exposure)
```

```python
print(photosynthesis)
```

```python
sample_size=50

plt.figure(figsize=(8,8))

dot_sizes=np.pi*(15*np.random.rand(sample_size)**2)
colors=np.random.rand(sample_size)

plt.scatter(photosynthesis, sunlight_exposure, marker='o', s=dot_sizes, c=colors)

plt.xlabel(r'sunlight exposure', fontsize=10)
plt.ylabel(r'photosynthesis', fontsize=10)

plt.title('Graph Representing Photosynthesis versus Sunlight Exposure \n', fontsize=10)

plt.savefig('graph_representing_photosynthesis_versus_sunlight_exposure.svg')

plt.show()
```

```python
plt.figure(figsize=(6,6))
    
num_bins = 100
x = np.random.randn(1000, 3)
    
colors = ['green', 'blue', 'lime']
  
plt.hist(x, num_bins, density=True, color = colors, label = colors)
  
plt.legend(loc="upper left", fontsize=12)
  
plt.show()
```

### Pie Charts

```python
%reset
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([35, 25, 25, 15]) # adds up to 100

fig = plt.figure(figsize=(5,5))

plt.pie(x)

plt.show()
```

```python
x = np.array([35, 25, 25, 15]) # adds up to 100

Fruit = ["Apples", "Bananas", "Cherries", "Dates"]

fig = plt.figure(figsize=(5,5))

plt.pie(x, labels = Fruit)

plt.show()
```

```python
y = np.array([35, 25, 25, 15]) # adds up to 100

Fruit = ["Apples", "Bananas", "Cherries", "Dates"]

fig = plt.figure(figsize=(5,5))

plt.pie(y, labels = Fruit, startangle = 90)

plt.show() 
```

```python
y = np.array([35, 25, 25, 15]) # adds up to 100

Fruit = ["Apples", "Bananas", "Cherries", "Dates"]

Segments = [0.2, 0.3, 0.4, 0.1]

fig = plt.figure(figsize=(5,5))

plt.pie(y, labels = Fruit, explode = Segments)

plt.show() 
```

```python
y = np.array([35, 25, 25, 15]) # adds up to 100

Fruit = ["Apples", "Bananas", "Cherries", "Dates"]

Segments = [0.1, 0.2, 0.3, 0.4]

fig = plt.figure(figsize=(5,5))

plt.pie(y, labels = Fruit, explode = Segments, shadow = True)

plt.show() 
```

```python
y = np.array([35, 25, 25, 15]) # adds up to 100

Fruit = ["Apples", "Bananas", "Cherries", "Dates"]

Colours = ["black", "hotpink", "b", "#4CAF50"]

fig = plt.figure(figsize=(5,5))

plt.pie(y, labels = Fruit, colors = Colours)

plt.show()
```

```python
y = np.array([35, 25, 25, 15]) # adds up to 100
 
Fruit = ["Apples", "Bananas", "Cherries", "Dates"]

fig = plt.figure(figsize=(8,8))

plt.pie(y, labels = Fruit)

plt.legend(title = "Fruits", loc="upper right")

plt.show() 
```
